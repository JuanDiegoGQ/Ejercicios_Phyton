***punto 1 SEÑAL CUASIPERIÓDICA***

$x(t) = 8\cos(100t) + 10\sin(20t) - 7\cos\left(300t - \frac{\pi}{7}\right)$


```python
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
ti=0
tf=(2*np.pi/10)
t=np.arange(ti,tf,0.001)
x= 8*np.cos(100*t)+10*np.sin(20*t)-7*np.cos(300*t-np.pi/7)
plt.plot(t,x)
plt.xlabel("Tiempo[t]")
plt.ylabel("X[t]")
plt.title("señal cuasiperiódica")
plt.grid(True)
plt.show()

y_min = 0 # valor min después de cero y pendiente
y_max = 5 # valor max después de cero y pendiente
#crear funcion en python
def my_ceropen(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
t_2=my_ceropen(x,0,5)
plt.plot(t,t_2)
plt.xlabel("Tiempo[t]")
plt.ylabel("X[t]")
plt.title("señal cuasiperiódica")

nbits=3
rmin=0
rmax=5
ve=np.linspace(rmin,rmax,2**nbits)

###vector de cuantización
d = cdist(t_2.reshape(-1,1),ve.reshape(-1,1))
d.shape
ind= np.argmin(d,axis=1)
ind
print(t_2[1],ind[1],ve[ind[1]])



```
