Ejerccio 1:
Use la función $rect(t)$ para construir una función $cos(ωt)$ con $ω=2$ la cual se enciende en $t=−1/2$ y se apaga en $t=+1/2.$

```python
import sympy as sym #importa la librería sympy
sym.init_printing() #se inicializa para mostrar en pantalla
t, sigma, omega = sym.symbols('t sigma omega', real=True)
x=sym.cos(omega*t)
x
```

```python


class rect(sym.Function): #clase/objeto tipo sympy

    @classmethod #decorador objeto
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)

sym.plot(x.subs(omega,2)*rect(t), (t, -5, 5), ylim=[0, 1], ylabel=r'rect$(t)$');
```
